```python
laliga.sort_values(["Position",'Games played'], ascending=False)
```

```python
laliga.sort_values(["Position",'Games played'], ascending=[True,False])
```

```python
rule = {"Goalkeeper": 1,"Defender": 2,"Midfielder": 3,"Forward": 4}
laliga.sort_values(by=["Position"], key=lambda x: x.apply(lambda y: rule.get(y, 10)))
```

```python
laliga.sort_values(by=["Position",'Games played'], key=lambda x: x.map(rule).fillna(x),ascending=[True,False])
```
